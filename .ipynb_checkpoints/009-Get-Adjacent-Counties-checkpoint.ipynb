```python
import requests as req
import time
DATE = time.strftime("%m-%d-%H_")
URL ="https://www2.census.gov/geo/docs/reference/county_adjacency.txt"
resp = req.get(URL)
content = resp.text
lineZZ = content.splitlines()
```

```python
import re
lines=lineZZ
text=''
cnt=0
for line in lines:
    cnt=cnt+1
    line = line.replace('"','')
    line = line.replace('\n','')
    line0 = ' '.join(line.split())
    linez = line.split(" ")
    if len(linez) ==5:
        text=text+"\n:"+linez[0]+":\n"
        text=text+"\n:"+line0+":\n"
    if len(linez) !=5:
        text=text+line0+"\n"
        
        
print(len(text))
textS = text.split("\n")
print(len(textS))        
 
TX =""
def replaceNth(s, source, target, n):
    inds = [i for i in range(len(s) - len(source)+1) if s[i:i+len(source)]==source]
    if len(inds) < n:
        return  # or maybe raise an error
    s = list(s)  # can't assign to string slices. So, let's listify
    s[inds[n-1]:inds[n-1]+len(source)] = target  # do n-1 because we start from the first occurrence of the string, not the 0-th
    return ''.join(s)

source = ' '
target = '\n'
n = 3
cnt=0
for line in textS:
    line=line.replace(" County, ",", ")    
    cnt=cnt+1
    lineZ =line.split(",")
    #print(len(lineZ))
    if len(lineZ)==3 and len(line)>5:# and line[-1]!=":":
        TX=TX+replaceNth(line, source, target, n)+"\n"
    if len(lineZ)!=3 and len(line)>5 and line[-1]!=":":
        TX=TX+line+"\n"
        
        
NTEXT =""
cnt = 0
TXs = TX.split("\n")
for line in TXs:
    if cnt<40:
        NTEXT=NTEXT+line.rstrip(":")+" "
        
COUNTIES =[]
lines = NTEXT.split(":")
for line in lines:
    if len(line)>5:
        COUNTIES.append([line])        
        
#test:
for i in range(0,len(COUNTIES)):   
    for lines in COUNTIES[i]:
        if "Manatee" in lines:
            print(COUNTIES[i])
            
np.save("resources/COUNTIES", COUNTIES) 
print("\nTEST:\n")
cnt=0
COUNTIES = np.load("resources/COUNTIES.npy")
for counties in COUNTIES:
    cnt=cnt+1
    if cnt<20:
        print(counties)
```

```python
import codecs
import string
cnt=0
file_name ="Custom/CountyAdjacency.py"
savfile=open(file_name,"w")
filename ="Custom/county_adjacency.py"
with codecs.open(filename, 'r', encoding='utf-8',errors='ignore') as fdata:
    for line in fdata:
        if len(line)>5:
            cnt=cnt+1
            line = line.replace("\n","")
            line = line.replace("\"","'")
            line=line.replace("\t"," ")
            #if line[0] =='"':line=":|"+line
            printable = set(string.printable)
            lines = ''.join(filter(lambda x: x in printable, line))
            #if lines[0] =='"':lines=":|"+lines
            items=lines.split(":")
            for item in items:
                item=item.replace("'''","\"\"\"")
                if item[0] =='\'':item=":|"+item
                #print(item)
                #lines = str(lines)
                #print(lines)
                if len(item)>5:
                     savfile.write(item+"\n")
                        
savfile.close()            
```
